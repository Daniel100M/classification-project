```python
# When we apply numbers to a machine learning algorithm, ultimately the cross entropy (the loss value) is a single number.
# Now, since the loss value is computed based on the model's performance, we can determine that the higher the loss value, the worse the model.
# That is because once the model achieves it's prediction values, the loss is calculated by comparing those values with the target values.
# And so, the bigger the difference between the target values and the model's actual predictions, the higher the loss values will be.

# If a certain feature has a very high range of values, it tends to dominate the loss.
# However, if a feature has a lower range of values, then the loss won't be affected much by it.
# Now, since what dominates the loss also dominates the optimization process(the adjusment of weights so that the model could perform better and achieve better predictions - more accurate probability values),
# features with greater ranges will have their rates changed much more drastically than features with smaller ranges.
# This will in turn affect our model's predictions badly.

# Thus, our objective would be leveling the playing field for all the numeric features in the data.
# Meaning we wish to avoid situations in which some features have a much greater numeric range (minimum value to maximum value) than others.
# This will assist in the optimization process as well as the computation of the loss value. Ultimately this will lead to a better Logistic Regression Model.
```

```python
# In order to achieve this balance, we'll scale all numeric features to the range (0,1).
```

#### Now, since machine learning models can only be trained with numeric data, we need to convert categorical data to numbers. To do so, we'll use a  common technique which will use one-hot encoding for categorical columns. In this method we check all the different categories and create a column for each one. Category column values will be either 1 or 0. 1, If the row is of that category, 0 otherwise.


```python
df[categorical_cols].isna().sum()
```

```python
# So, for example, since there are 49 different locations,
# we'd like to create a column for each such location, and
# insert 0's and 1's accordingly.
```

```python
# Let's import 'OneHotEncoder' in order to do so, from the 
# sklearn.preprocessing model.
from sklearn.preprocessing import OneHotEncoder
```

```python
# First, as we have done in a similar way before, let's create
# a 'OneHotEncoder' object:
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
```

```python
# sparse = False : Because we don't want to create a sparse matrix, we want to create a numpy array.
# handle_unknown = 'ignore' : If we come across a category that we haven't seen before- ignore it.
```

```python
# Now, since the encoder needs to first understand the data,
# we will once again use the 'fit' function.
# Identify the full list of categories across all categorical columns.
encoder.fit(df[categorical_cols])
```

```python
# Now let's view which categories the encoder
# has identified:
encoder.categories_
```

```python
# What we'd like to do is generate a whole column for each one
# of the values above. To do so, we'll use a function called:
# 'get_feature_names'
encoded_cols = list(encoder.get_feature_names(categorical_cols))
print(encoded_cols)
```

```python
# As we can see, we now have a list on each and every single
# category value.
```

```python
# Now we'd like to generate those columns and add them to the
# data-frame. To do so, we'll use the 'transform' method
# (which will perform the encoding) on each set:
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

```python
# In order to train a Logistic Regression model, we'll use
# the 'LogisticRegression' class from the Scikit-learn 
# library.
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Now, let's create a model object:
model = LogisticRegression(solver='liblinear')
```

```python
# We can clearly see that there are many different 
# parameters we can pass to the model.
# Now, each time we try a different kind of model, with
# a different set of parameters, we could use the validation
# set to determine whether these changes have led to a
# better result.
```

```python
# The optimization technique we're using is called lib-linear.

# The parameter 'max_iter' sets the amount of iterations of
# our model. Meaning how many times the model will continue
# the Logistic Regression Workflow. We could set it to any 
# numeric value, however the default is set to 100.

# The parameter 'tol' represents the tolerance. Meaning it
# represents a number, the default of which is 0.001, which,
# once the loss value is less than or equal to this value
# the model will stop the training process.
```

```python
# In order to train the model, we'll once again use the '.fit'
# function:
model.fit(train_inputs[numeric_cols + encoded_cols], train_targets)
```

```python
# The code above trains our model and uses the 
# Logistic Regression training workflow:
#
# 1) We insert into the model the training input, including
#    the target outputs of the training.
# 2) We then take the model's output (the prediction), and we
#    compare it to the actual target values.
#
# According to the model's prediction success:
#
# 3) We calculate a loss value using a loss function (higher loss = bad model prediction, lower loss = good model prediction).
# 4) Afterwards, we use an optimization technique to slightly reduce the loss and improve 
#    The weights of the model.
# 5) Rince and repeat these steps until the model's predictions are good enough.
```

```python
# Notice that we're sending the '.fit' function two arguments,
# which are 'train_inputs[numeric_cols + encoded_cols]' and
# 'train_targets'.

# Now, since the model can only process numeric values, note
# that we specify 'numeric_cols + encoded_cols' which will 
# adress the training input columns which contain numeric 
# values.

# However, 'train_targets' is a pandas series of two categorical
# values: 'Yes'\'No'. Fortunately, scikit-learn's Logistic
# Regression model can work with categorical values
# once there are only two of them. In which case, 
# the first class the model encounters will be converted to 0
# and the second to 1. 
```

```python
print(numeric_cols + encoded_cols)
```

```python
# Let's view the coefficients that have been assigned to 
# each training feature.
print(model.coef_.tolist())
```

```python
# Let's quickly create a data-frame that will showcase
# the coefficient (weight)
# that has been assigned to each feature.
pd.DataFrame({
    'feature': (numeric_cols + encoded_cols),
    'weight': model.coef_.tolist()[0]
})
```

```python
# As we can see above,

# 'MinTemp' has been assigned a coefficient of 0.9
# which represents the weight of the feature numeric value
# of 'MinTemp'.
# This weight is quite close to one, which is good, because
# it indicates that the lower the 'MinTemp' is, the more
# likely it is going to rain tomorrow.

# 'MaxTemp' has been assigned a coefficient of -2.9
# Which means that the maximum temperature has a negative
# affect on whether it is going to rain tomorrow. This
# basically indicates that the higher the maximum temperature,
# the less likely it will rain tomorrow.

# 'RainFall' has been assigned a coefficient (weight) of
# 3.16, which makes sense because the more rain-fall we had
# today, the higher the probability of it raining tomorrow.
```

```python
# To sum up, the sign of the weight indicates whether
# it is a positively correlated feature or negatively
# correlated feather.
# Additionally, the abstract value of the weight indicates it's
# importance in predicting whether it will rain tomorrow.
```

```python
# The added weight at the calculation
# which is a 'zero-mean-correction'.
print(model.intercept_)
```
